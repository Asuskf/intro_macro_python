**Fianlizamos:** Automatizando los pasos

```python
import pandas as pd
```

```python
# leer
pib_df = pd.read_excel("/content/PIB.xls", skiprows=3)
pib_df
```

```python
# filtrar
ec = pib_df[pib_df["Country Name"] == "Ecuador"]
ec
```

```python
# graficar
columns_name = list(ec.columns)
index_columns = columns_name.index("Indicator Code")
transposed_df = ec[columns_name[index_columns+1:]].T
transposed_df.columns = ['PIB']
transposed_df.plot(title="PIB Ecuador")
```

```python
# Automatizar
def graph_gdp(country_name:str, dataframe: pd.DataFrame)->None:
    """
    Graph the GDP of a country

    Args:
        country_name (str): Country name to analyze
        dataframe (pd.Dataframe): All data of GDP

    Returns:
        _type_: None
    """
    aux_df = dataframe[dataframe["Country Name"] == country_name]
    columns_name = list(aux_df.columns)
    index_columns = columns_name.index("Indicator Code")
    transposed_df = aux_df[columns_name[index_columns+1:]].T
    transposed_df.columns = ['PIB']
    transposed_df.plot(title=f"PIB {country_name}")
    return None
```

```python
graph_gdp("Colombia", pib_df)
```

Usar algunas caraterísticas de **Colab**

```python
import ipywidgets as widgets
from IPython.display import display
# Create the dropdown widget
dropdown = widgets.Dropdown(options=pib_df["Country Name"].dropna(), description='Country:')
# Create an output area to display information
output = widgets.Output()
# Define a function to react to dropdown changes
def on_dropdown_change(change):
    selected_country = change.new
    country_data = pib_df[pib_df['Country Name'] == selected_country]

    with output:
        output.clear_output()
        graph_gdp(selected_country, pib_df)
        print(f"Selected Country: {selected_country}")

# Attach the function to observe dropdown changes
dropdown.observe(on_dropdown_change, names='value')

# Display the dropdown and the output area
display(dropdown)
display(output)
```
