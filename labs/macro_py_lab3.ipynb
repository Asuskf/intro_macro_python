# Lab3
**Objetivo:** Hacer una representación gráfica de consumo e ingreso disponible

```python
import matplotlib.pyplot as plt
import numpy as np

# Función de consumo
def calcular_consumo(yd, c0, c1)-> float:
    """
    Calcula el consumo utilizando la función de consumo del modelo keynesiano.

    Parámetros:
    yd (float): Ingreso disponible después de impuestos.
    c0 (float): Componente autónomo del consumo.
    c1 (float): Propensión marginal al consumo.

    Returns:
    float: El nivel de consumo calculado.
    """
    consumo = c0 + c1 * yd
    return consumo

# Parámetros de la función de consumo
c0 = 50
c1 = 0.6

# Valores de YD (ingreso disponible)
yd_values = np.linspace(0, 200, 100)


# Calcular el consumo correspondiente a cada valor de YD
consumo_values = [calcular_consumo(yd, c0, c1) for yd in yd_values]

# Crear la gráfica
plt.figure(figsize=(8, 6))
plt.plot(yd_values, consumo_values, label=f'Función de Consumo: C = {c0} + {c1} YD')
plt.xlabel('Ingreso Disponible (YD)')
plt.ylabel('Consumo (C)')
plt.title('Relación entre Consumo e Ingreso Disponible')
plt.legend()
plt.grid(True)
plt.show()
```
