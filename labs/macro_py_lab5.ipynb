Ejercicio 1: Calcular el Consumo

Dados los valores:

$(c_0 = 100$),

$(c_1 = 0.6$),

$(Y = 500$),

$(T = 50$).

Calcular el nivel de consumo utilizando la fórmula:

$[ C = c_0 + c_1 (Y - T) $]


**Ejercicio 2:**

Cambio en el Consumo por Cambio en los Impuestos Utilizando los mismos valores
$(c_0$) y $(c_1$) del ejercicio anterior, considerar que: $(Y = 800$) y $(T$) aumenta a $(100$).

Calcular el cambio en el consumo debido al aumento en los impuestos, utilizando la fórmula:

[ Cambio en el consumo = $C_2 - C_1 $]

donde $(C_1$) es el consumo inicial y $(C_2$) es el consumo con el nuevo valor de los impuestos.


```python
import matplotlib.pyplot as plt
import numpy as np

# Valores del Ejercicio 1
c0 = 100
c1 = 0.6
Y = 500
T = 50

# Rango de valores para los impuestos (T)
T_range = np.linspace(0, 200, 100)

# Cálculo del consumo utilizando la fórmula para el Ejercicio 1
consumo_ejercicio1 = c0 + c1 * (Y - T_range)

# Nuevos valores para el Ejercicio 2
Y_nuevo = 800
T_nuevo = T + 100

# Cálculo del consumo con los nuevos valores para el Ejercicio 2
consumo_nuevo_ejercicio2 = c0 + c1 * (Y_nuevo - T_range)

# Rango de valores para los ingresos totales (Y)
Y_range = np.linspace(0, 1000, 100)

# Cálculo del consumo utilizando la fórmula para el Ejercicio 1
consumo_ejercicio1_y = c0 + c1 * (Y_range - T)

# Cálculo del consumo con los nuevos valores para el Ejercicio 2
consumo_nuevo_ejercicio2_y = c0 + c1 * (Y_range - T_nuevo)

# Gráfico 2D Consumo vs. Impuestos
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.plot(T_range, consumo_ejercicio1, label=r'$C = c_0 + c_1 (Y - T)$ (Ejercicio 1)')
plt.plot(T_range, consumo_nuevo_ejercicio2, label=r'$C = c_0 + c_1 (Y - \hat{T})$ (Ejercicio 2)')
plt.scatter([T], [c0 + c1 * (Y - T)], color='red', marker='o', label='Punto Inicial')
plt.scatter([T_nuevo], [c0 + c1 * (Y_nuevo - T_nuevo)], color='green', marker='o', label='Nuevo Punto')
plt.xlabel('Impuestos (T)')
plt.ylabel('Consumo (C)')
plt.title('Consumo e Impuestos')
plt.axvline(x=T, color='gray', linestyle='--', label='T Inicial')
plt.axvline(x=T_nuevo, color='orange', linestyle='--', label='Nuevo T')
plt.axhline(y=c0 + c1 * (Y - T), color='gray', linestyle='--')
plt.axhline(y=c0 + c1 * (Y_nuevo - T_nuevo), color='orange', linestyle='--')
plt.legend()

# Vector de Desplazamiento
plt.arrow(T, c0 + c1 * (Y - T), T_nuevo - T, c0 + c1 * (Y_nuevo - T_nuevo) - (c0 + c1 * (Y - T)), head_width=10, head_length=15, fc='blue', ec='blue')
plt.text((T + T_nuevo) / 2, (c0 + c1 * (Y - T) + c0 + c1 * (Y_nuevo - T_nuevo)) / 2, 'Desplazamiento', color='blue', ha='center')

plt.grid(True)

# Gráfico de Barras Comparativo
plt.subplot(1, 3, 2)
nombres = ['Consumo Inicial', 'Nuevo Consumo']
valores = [c0 + c1 * (Y - T), c0 + c1 * (Y_nuevo - T_nuevo)]
plt.bar(nombres, valores, color=['blue', 'green'])
plt.xlabel('Escenario')
plt.ylabel('Consumo')
plt.title('Comparación de Consumos')
plt.ylim(0, max(valores) + 50)
plt.grid(axis='y')

# Gráfico 2D Consumo vs. Ingresos Totales
plt.subplot(1, 3, 3)
plt.plot(Y_range, consumo_ejercicio1_y, label=r'$C = c_0 + c_1 (Y - T)$ (Ejercicio 1)')
plt.plot(Y_range, consumo_nuevo_ejercicio2_y, label=r'$C = c_0 + c_1 (Y - \hat{T})$ (Ejercicio 2)')
plt.scatter([Y], [c0 + c1 * (Y - T)], color='red', marker='o', label='Punto Inicial')
plt.scatter([Y_nuevo], [c0 + c1 * (Y_nuevo - T_nuevo)], color='green', marker='o', label='Nuevo Punto')
plt.xlabel('Ingreso Total (Y)')
plt.ylabel('Consumo (C)')
plt.title('Consumo vs. Ingresos Totales')
plt.axvline(x=Y, color='gray', linestyle='--', label='Y Inicial')
plt.axvline(x=Y_nuevo, color='orange', linestyle='--', label='Nuevo Y')
plt.axhline(y=c0 + c1 * (Y - T), color='gray', linestyle='--')
plt.axhline(y=c0 + c1 * (Y_nuevo - T_nuevo), color='orange', linestyle='--')
plt.legend()

# Vector de Desplazamiento
plt.arrow(Y, c0 + c1 * (Y - T), Y_nuevo - Y, c0 + c1 * (Y_nuevo - T_nuevo) - (c0 + c1 * (Y - T)), head_width=20, head_length=30, fc='blue', ec='blue')
plt.text((Y + Y_nuevo) / 2, (c0 + c1 * (Y - T) + c0 + c1 * (Y_nuevo - T_nuevo)) / 2, 'Desplazamiento', color='blue', ha='center')

plt.grid(True)

# Ajustar espaciado entre subplots
plt.tight_layout()

# Mostrar los gráficos
plt.show()

# Imprimir resultados
print("Resultados:")
print(f"Consumo Inicial (Ejercicio 1): {c0 + c1 * (Y - T):.2f}")
print(f"Nuevo Consumo (Ejercicio 2): {c0 + c1 * (Y_nuevo - T_nuevo):.2f}")
print(f"Cambio en el Consumo: {c0 + c1 * (Y_nuevo - T_nuevo) - (c0 + c1 * (Y - T)):.2f}")
```
